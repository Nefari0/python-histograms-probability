```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
shift = 5
stretch = 2
n = 1000


pnts = np.random.rand(n)*stretch + shift-.5*stretch
print('Mean is %g, range is %g' %(np.mean(pnts),np.max(pnts)-np.min(pnts)))


# pnts = np.random.randn(n)*stretch + shift
# print('Mean is %g, Standard Deviation is %g' %(np.mean(pnts),np.std(pnts)))

fig,ax = plt.subplots(1,2,figsize=(6,4))

ax[0].plot(pnts,'s',alpha=.4)
# ax[1].hist(pnts,50)
ax[1].hist(pnts,50,edgecolor='k')

plt.show()
```

```python
lam = 4
n = 1000


pnts = np.random.poisson(lam,n)*stretch + shift-.5*stretch
print('Mean is %g, variance is %g' %(np.mean(pnts),np.var(pnts)))

fig,ax = plt.subplots(1,2,figsize=(6,3))

ax[0].plot(pnts,'s',alpha=.4)
# ax[1].hist(pnts,50)
ax[1].hist(pnts,bins=np.arange(0,np.max(pnts) + 1),edgecolor='k')

plt.show()
```
